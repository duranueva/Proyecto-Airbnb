```python
import requests

url = "https://raw.githubusercontent.com/duranueva/Proyecto-Airbnb/refs/heads/main/tools/listings.csv"
file_name = "listings.csv"

response = requests.get(url)
if response.status_code == 200:
    with open(file_name, "wb") as file:
        file.write(response.content)
    print("Archivo descargado correctamente")
else:
    print(f"Fallo al descargar el archivo. Status: {response.status_code}")
```

# Creating table

```python
columns = list(airbnb_df.columns)
columns
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('airbnb.db')
cursor = conn.cursor()
create_table_sql = """
CREATE TABLE IF NOT EXISTS Mexico_city (
    id INTEGER,
    name TEXT,
    host_id INTEGER,
    host_name TEXT,
    neighbourhood TEXT,
    latitude DECIMAL(9, 6),  -- Great for locations
    longitude DECIMAL(9, 6),
    room_type TEXT,
    price FLOAT,
    minimum_nights INTEGER,
    number_of_reviews INTEGER,
    last_review DATE,
    reviews_per_month FLOAT,
    calculated_host_listings_count INTEGER,
    availability_365 INTEGER,
    number_of_reviews_ltm INTEGER
);
"""
cursor.execute(create_table_sql)
conn.commit()
conn.close()
```

# Inserting Data

```python
def isNaN(value):
  if pd.isna(value):
    return 'NULL'
  else:
    if isinstance(value, str):
      if "'" in value:
        value = value.replace("'", "''")

      return "'" +value+ "'"
    else:
      return value
```

```python
connection = sqlite3.connect("airbnb.db")
cursor = connection.cursor()

# Sentencia SQL, hacemos un f-string multilinea
for i in range(0, len(airbnb_df)):
  sql_insertion = f"""
    INSERT INTO
      Mexico_city
    VALUES(
      -- 'var',   -- TEXT o DATE
      --  var,    -- Cualquier otro tipo de Datos
      {isNaN(airbnb_df.loc[i, "id"])},
      {isNaN(airbnb_df.loc[i, "name"])},
      {isNaN(airbnb_df.loc[i, "host_id"])},
      {isNaN(airbnb_df.loc[i, "host_name"])},
      {isNaN(airbnb_df.loc[i, "neighbourhood"])},
      {isNaN(airbnb_df.loc[i, "latitude"])},
      {isNaN(airbnb_df.loc[i, "longitude"])},
      {isNaN(airbnb_df.loc[i, "room_type"])},
      {isNaN(airbnb_df.loc[i, "price"])},
      {isNaN(airbnb_df.loc[i, "minimum_nights"])},
      {isNaN(airbnb_df.loc[i, "number_of_reviews"])},
      {isNaN(airbnb_df.loc[i, "last_review"])},
      {isNaN(airbnb_df.loc[i, "reviews_per_month"])},
      {isNaN(airbnb_df.loc[i, "calculated_host_listings_count"])},
      {isNaN(airbnb_df.loc[i, "availability_365"])},
      {isNaN(airbnb_df.loc[i, "number_of_reviews_ltm"])}
    )
  """
  cursor.execute(sql_insertion) # Execute sentence
  print("Incersion " + str(i+1) + " exitosa")
  connection.commit() # Upload the changes
connection.close() # Closes the connection
```

# Analisis Exploratorio

```python
def query(sentence):
  conn = sqlite3.connect('airbnb.db')
  result = pd.read_sql(sentence, conn)
  conn.close()
  print(result)
```

Observemos el numero de registros o lugares que Airbnb tiene registrados

```python
sql_query = """
SELECT COUNT(*) FROM Mexico_city
"""

query(sql_query)
```

Observemos los distintos vecindarios que Airbnb tiene y el numero de lugares cada uno, del mayor al menor.

```python
sql_query = """
SELECT neighbourhood, COUNT(neighbourhood) as total_places FROM Mexico_city
GROUP BY neighbourhood
ORDER BY total_places DESC
"""

query(sql_query)
```

Observemos los distintos tipos de estancias y el numero de estas, de mayor al menor.

```python
sql_query = """
SELECT room_type, COUNT(room_type) as total_places FROM Mexico_city
GROUP BY room_type
ORDER BY total_places DESC
"""

query(sql_query)
```

Observemos la cantidad de anfitriones respecto a su host_id

```python
sql_query = """
SELECT COUNT(DISTINCT(host_id)) as Number_of_unique_hosts FROM Mexico_city
"""

query(sql_query)
```

Observemos los hosts con el mayor numero de propiedades

```python
sql_query = """
SELECT host_name, COUNT(host_id) as total_places FROM Mexico_city
GROUP BY host_id
ORDER BY total_places DESC
LIMIT 15
"""

query(sql_query)
```

```python
sql_query = """
SELECT host_name, COUNT(*) as total_places FROM Mexico_city
WHERE host_name = "Mr. W"
"""

query(sql_query)
```
